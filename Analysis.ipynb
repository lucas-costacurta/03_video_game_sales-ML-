```python
import pandas as pd
import graphviz
import pydotplus
from IPython.display import Image
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

music_data = pd.read_csv('music.csv')
x = music_data.drop(columns=['genre'])
y = music_data['genre']

model = DecisionTreeClassifier()
model.fit(x,y)
prediction = model.predict([[21,1]])
prediction[0]
```

```python
# simple prediction

gender = int(input('Qual o sexo dessa pessoa? (0 - Feminino; 1 - Masculino)'))
age = int(input('Qual a idade dessa pessoa?'))

prediction = model.predict([[age,gender]])
print(f'Uma pessoa do sexo {'masculino' if gender == 1 else 'feminino'}, de {age} anos, pode preferir o estilo {prediction[0]}')
```

```python
# accuracy score

music_data = pd.read_csv('music.csv')
x = music_data.drop(columns=['genre'])
y = music_data['genre']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

model = DecisionTreeClassifier()
model.fit(x_train,y_train)
prediction = model.predict(x_test)

score = accuracy_score(y_test, prediction)
score
```

```python
# visualizing the decision tree

music_data = pd.read_csv('music.csv')
x = music_data.drop(columns=['genre'])
y = music_data['genre']

model = DecisionTreeClassifier()
model.fit(x,y)

graph_tree = export_graphviz(model, out_file='music_recommender.dot',
                feature_names=['age','gender'],
                class_names=sorted(y.unique()),
                rounded=True,
                filled=True)

plot_graph = graphviz.Source(graph_tree)
plot_graph.render(filename='iris_tree', 
             format='png', 
             view=True)
```
